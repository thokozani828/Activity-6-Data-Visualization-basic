```python
# Extracting data
player_names = df['Name']
goals_scored = df['Goals']
player_age = df['Age']

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(player_age, goals_scored, color='blue', alpha=0.5)
plt.title('Goals Scored vs. Age')
plt.xlabel('Age')
plt.ylabel('Goals Scored')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
player_age = df['Age']

# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(player_age, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Extracting player names and goals scored
player_names = df['Name']
goals_scored = df['Goals']

# Plotting bar chart
plt.figure(figsize=(12, 6))
plt.bar(player_names, goals_scored, color='skyblue')
plt.title('Goals Scored by Players')
plt.xlabel('Player Name')
plt.ylabel('Goals Scored')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Count the number of players for each position
position_counts = df['Position'].value_counts()

# Plotting pie chart
plt.figure(figsize=(8, 8))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', colors=plt.cm.tab10.colors, startangle=140)
plt.title('Distribution of Players by Position')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
